```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
from langchain_ollama import ChatOllama
```

```python
llm = ChatOllama(
    model = 'mistral'
)
```

```python
class BlogState(TypedDict):
    title: str
    outline: str
    content: str    
```

```python
def generate_outline(state: BlogState) -> BlogState:
    #fetch title
    title = state['title']

    #generate outline
    prompt = f'Generate a brief outline for a blog on the topic titled - {title}'
    outline = llm.invoke(prompt)

    #update state
    state['outline'] = outline
    return state  
```

```python
def generate_content(state: BlogState) -> BlogState:
    #fetch title and outline
    title = state['title']
    outline = state['outline']

    #generate content
    prompt = f'Write a small blog on the title - {title} using the outline \n {outline}'

    #prompt the llm
    content = llm.invoke(prompt).content

    #update content
    state['content'] = content
    return state
```

```python
graph = StateGraph(BlogState)

#add nodes
graph.add_node('generate_outline', generate_outline)
graph.add_node('generate_content', generate_content)

#add edges
graph.add_edge(START, 'generate_outline')
graph.add_edge('generate_outline', 'generate_content')
graph.add_edge('generate_content', END)

#compile the graph
workflow = graph.compile()
```

```python
#define initial state
initial_state = {'title': 'Use cases of langgraph'}

#start the workflow
final_state = workflow.invoke(initial_state)

print(final_state['outline'])
print(final_state['content'])
```
